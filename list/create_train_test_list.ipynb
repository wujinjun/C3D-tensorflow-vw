```python
#pytrain.listtest.list
#
#create_train_test_list("/home/wjj/HDD/HDD/UCF-101/","train.list","test.list")
#0.750
import os
def create_train_test_list(basepath,train_list,test_list,video_number=200,train_factor=0.7):    #basepathinputlist outputprefix
    f_train_list = open(train_list,'w')    #
    f_test_list = open(test_list,'w')    #
    classlist =  os.listdir(basepath)    #basepathclasslist
    label = 0  
    classlist.sort()    #
    for classvar in classlist:    #for    
        videodir = os.path.join('%s%s' % (basepath,classvar))
        if os.path.isdir(videodir) == True:    #
            sample_number = 0
            videolist = os.listdir(videodir)    #videovideolist
            videolist.sort()
            #number_of_samples = len(videolist)
            number_of_samples = video_number        #
            train_samples_number = int(train_factor *  number_of_samples)    #
            #test_samples_number = int((1 - train_factor) * number_of_samples)    #
            #print(train_samples_number,test_samples_number)
            videolist.sort()
            for videovar in  videolist:    #forvideo
                clipdir = os.path.join('%s/%s' % (videodir,videovar))
                if os.path.isdir(clipdir) == True:
                    sample_number += 1
                    #print(clipdir)
                    if sample_number <= train_samples_number:    #train
                        print("train",clipdir,label)
                        f_train_list.write('%s %s\n' % (clipdir,label))
                    else :
                        #print("test",clipdir,label)
                        f_test_list.write('%s %s\n' % (clipdir,label))
                    if sample_number >= number_of_samples:
                        break                               
            label=label+1
    f_test_list.close()        
    f_train_list.close()

create_train_test_list("/samba/HDD/UCF-10/","train_UCF10.list","test_UCF10.list",100,0.7)
#create_train_test_list("/home/wjj/HDD/HDD/UCF-101/","train.list","test.list",[],[])
```
